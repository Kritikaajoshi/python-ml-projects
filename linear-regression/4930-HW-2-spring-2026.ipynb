#### Step 5.2: Encode with `OneHotEncoder`
* We will use `OneHotEncoder(handle_unknown="ignore")` to encode the nominal columns.
* Nominal categories have no natural order. For instance, take a look at values in the `HISPMOM` column:

```
C:Cubans, M:Mexicans,  N:No, O:Colombians P:Peruvians, S:Salvadorans, U:Guatemalans
```
if encoded as 

```
0:Cubans, 1:Mexicans,  2:No, 3:Colombians 4:Peruvians, 5:Salvadorans, 6:Guatemalans
```
The model may incorrectly assume:

```
Cubans < Mexicans < No < Colombians < Peruvians < Salvadorans < Guatemalans

or, 

C < M < N < O < P < S< U
```
That is mathematically incorrect.
Instead, one-hot encoding produces:

| HISPMOM_C | HISPMOM_M | HISPMOM_N | HISPMOM_O | HISPMOM_P | HISPMOM_S | HISPMOM_U |
| ----------- | ----------- | ------------ |------------ |------------ |------------ |------------ |
| 1           | 0           | 0            |0            |0            |0            |0            |

* As you can see, each category gets its own binary column.
* If test data contains a category not seen in training, the option `handle_unknown="ignore"` in the `OneHotEncoder` will encode all one-hot columns for the nominal variable become 0 instead of causing an error. This prevents the model from crashing.

````python
#Step 5: Nominal column transformer

from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

nominal_transformer = Pipeline(
	steps=[] #List of tuples of each step in this pipeline: (), (), ...
)

# YOUR CODE HERE
nominal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

````

* **Task 4.3**: Complete the **RMSE()** function: 
  * It takes two lists: `y_test` ($y$), `y_pred` ($\hat{y}$), where the first list represents ground truth (i.e., actual) target values for the given samples, and the second list represents a corresponding predicted values for exactly same number of samples in `y_test`. 
  * Compute and return the `Root Mean Squared Error (RMSE)` of the prediction:

$$
\text{RMSE}(\hat{y}, y) = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

* In tasks 4.1, 4.2, and 4.3, PLEASE DO NOT USE ANY LIBRARY FUNCTION (e.g., scikit-learn) THAT DOES THE LINEAR REGRESSION FOR YOU or RMSE calculation.


````python
# Task 4.1: Complete the following function

# YOUR CODE HERE
def linear_regression_closed_form_training(X_train, y_train):
    # Add a column of ones for the intercept term (beta_0) 
    X_b = np.c_[np.ones((X_train.shape[0], 1)), X_train]
    
    # Calculate X^T * X 
    XTX = X_b.T @ X_b
    
    # Add a tiny value to the diagonal to prevent Singular Matrix error 
    # This is a standard way to handle collinearity in datasets 
    XTX_stable = XTX + 1e-5 * np.eye(XTX.shape[0])
    
    # Normal Equation: beta = (X^T * X)^-1 * X^T * y [cite: 383, 387]
    betas = inv(XTX_stable) @ X_b.T @ y_train
    return betas

    
    
````

````python
# YOUR CODE HERE
def linear_regression_closed_form_predict(betas, X_test):
    # Add intercept column to test data
    X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
    # Prediction: y = X * beta
    y_pred = X_b @ betas
    return y_pred
````

````python
#Task 4.3: Complete the following function
# YOUR CODE HERE
def RMSE(y_test, y_pred):
    # Convert inputs to numpy arrays to handle list inputs from tests
    y_test_arr = np.array(y_test)
    y_pred_arr = np.array(y_pred)
    
    # Compute Root Mean Squared Error
    rmse_val = np.sqrt(np.mean((y_test_arr - y_pred_arr)**2))
    return rmse_val
````

Make sure you use average of gradient in beta update formulas. Please use the `alpha` (i.e, the learning rate) and `nEpoch` (number of epochs) parameters in your implementation of the gradient descent algorithm.
  * The function returns the solution `beta`s (i.e., the `m+1` $\beta$ values).
* **Task 5.2**: Complete the **linear_regression_full_batch_gd_predict()** function:
  * It takes a list of `m+1` beta values (i.e., `beta`s returned from the corresponding training function, and `X_test_transformed` (containing test samples each having `m` input variables). 
  * Then, using the provided beta values, predict each of the test samples, and let's name your predictions `y_pred` ($\hat{y}$). Return `y_pred` ($\hat{y}$) from the function.

* In tasks 5.1, and 5.2, PLEASE DO NOT USE ANY LIBRARY FUNCTION (e.g., scikit-learn) THAT DOES THE LINEAR REGRESSION FOR YOU.

````python
#Task 5.1: Complete the function
# YOUR CODE HERE
def linear_regression_full_batch_gd_training(X_train, y_train, alpha, nEpoch):
    random.seed(554433)
    m, n = X_train.shape
    # Add intercept column of ones
    X_b = np.c_[np.ones((m, 1)), X_train]
    # Initialize betas to zeros
    betas = np.zeros(n + 1)
    
    for epoch in range(nEpoch):
        # Calculate gradients using the average gradient formula
        # Gradient = (2/m) * X^T * (X*beta - y)
        gradients = (2/m) * X_b.T @ (X_b @ betas - y_train)
        # Update betas
        betas = betas - alpha * gradients
        
    return betas
````

## TASK 6: Linear Regression with Stochastic Gradient Descent

* **Task 6.1**: Complete the **linear_regression_stochastic_gd_training()** function:

Make sure you use average of gradient in beta update formulas. Please use the `alpha` (i.e, the learning rate) and `nEpoch` (number of epochs) parameters in your implementation of the gradient descent algorithm.
  * The function returns the solution `beta`s (i.e., the `m+1` $\beta$ values).
* **Task 6.2**: Complete the **linear_regression_stochastic_gd_predict()** function:
  * It takes a list of `m+1` beta values (i.e., `beta`s returned from the corresponding training function, and `X_test_transformed` (containing test samples each having `m` input variables). 
  * Then, using the provided beta values, predict each of the test samples, and let's name your predictions `y_pred` ($\hat{y}$). Return `y_pred` ($\hat{y}$) from the function.

* In tasks 6.1, and 6.2, PLEASE DO NOT USE ANY LIBRARY FUNCTION (e.g., scikit-learn) THAT DOES THE LINEAR REGRESSION FOR YOU.


````python
# Task 6.1
def linear_regression_stochastic_gd_training(X_train, y_train, alpha, nEpoch):
    random.seed(554433)
    m, n = X_train.shape
    # Add intercept column
    X_b = np.c_[np.ones((m, 1)), X_train]
    betas = np.zeros(n + 1)
    
    for epoch in range(nEpoch):
        # Shuffling at the start of each epoch for better convergence
        indices = np.random.permutation(m)
        for i in indices:
            xi = X_b[i:i+1]
            yi = y_train.iloc[i:i+1]
            
            # Using the average-style gradient update for stability [cite: 505]
            # beta = beta - alpha * (2/m) * xi^T * (xi*beta - yi)
            prediction = xi @ betas
            gradient = (2.0 / m) * xi.T @ (prediction - yi)
            
            betas = betas - alpha * gradient.flatten()
            
    return betas
````

* **Task 7.1**: Complete the **linear_regression_minibatch_gd_training()** function:
  * It fits a linear regression model using the `Mini-batch gradient descent` to obtain the coefficients, beta's (i.e., $\beta$'s), as a numpy array of `m+1` values (Please recall class lecture), where $m$ is the number of variables in `X_train_transformed` (the first argument to the function). Make sure you use average of gradient in beta update formulas. Please use the `alpha` (i.e, the learning rate) and `nEpoch` (number of epochs) parameters in your implementation of the gradient descent algorithm.
  * The function returns the solution `beta`s (i.e., the `m+1` $\beta$ values).
* **Task 7.2**: Complete the **linear_regression_minibatch_gd_predict()** function:
  * It takes a list of `m+1` beta values (i.e., `beta`s returned from the corresponding training function, and `X_test_transformed` (containing test samples each having `m` input variables). 
  * Then, using the provided beta values, predict each of the test samples, and let's name your predictions `y_pred` ($\hat{y}$). Return `y_pred` ($\hat{y}$) from the function.

* In tasks 7.1, and 7.2, PLEASE DO NOT USE ANY LIBRARY FUNCTION (e.g., scikit-learn) THAT DOES THE LINEAR REGRESSION FOR YOU.


````python
# Task 7.1: Complete the following function
def linear_regression_minibatch_gd_training(X_train, y_train, alpha, nEpoch, batch_size):
    random.seed(554433)
    m, n = X_train.shape
    X_b = np.c_[np.ones((m, 1)), X_train]
    betas = np.zeros(n + 1)
    
    for epoch in range(nEpoch):
        # Shuffle indices to ensure random batches
        shuffled_indices = np.random.permutation(m)
        X_b_s = X_b[shuffled_indices]
        y_s = y_train.iloc[shuffled_indices]
        
        for i in range(0, m, batch_size):
            xi = X_b_s[i:i+batch_size]
            yi = y_s[i:i+batch_size]
            
            # Gradient update using the mini-batch average
            gradients = (2/len(xi)) * xi.T @ (xi @ betas - yi)
            betas = betas - alpha * gradients
            
    return betas
````

````python
# Task 7.2
def linear_regression_minibatch_gd_predict(betas, X_test):
    X_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
    y_pred = X_b @ betas
    return y_pred

````
